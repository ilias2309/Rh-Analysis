# Comprehensive Analysis of AI in HR Survey Data

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
file_path = 'processed_data.csv'
data = pd.read_csv(file_path)

# Define new column names for better comprehension
new_column_names_29 = [
    'Timestamp',
    'Age_Group',
    'Gender',
    'Current_Role',
    'Industry',
    'HR_Experience_Years',
    'AI_Implemented_in_HR',
    'Types_of_AI_in_HR',
    'AI_Integration_Level_in_HR',
    'AI_Benefits_in_HR',
    'AI_Challenges_in_HR',
    'AI_Tools_in_Recruitment',
    'Types_of_AI_in_Recruitment',
    'AI_Tools_Effectiveness_in_Recruitment',
    'Benefits_of_AI_in_Recruitment',
    'AI_Tools_in_Performance_Management',
    'Types_of_AI_in_Performance_Management',
    'AI_Tools_Effectiveness_in_Performance_Management',
    'Benefits_of_AI_in_Performance_Management',
    'Recommendations_for_AI_HR_Practices',
    'Satisfaction_with_AI_in_HR',
    'Improvements_for_AI_in_HR',
    'Unnamed1',
    'Unnamed2',
    'Unnamed3',
    'Unnamed4',
    'Unnamed5',
    'Unnamed6',
    'Unnamed7'
]

# Apply the new column names
data.columns = new_column_names_29

# Display the first few rows to verify the data
data.head()
    
```

## Descriptive Statistics

```python

# Generate descriptive statistics for numerical columns
descriptive_stats = data.describe()

# Display the descriptive statistics
descriptive_stats
    
```

## Distribution of Key Categorical Variables

```python

# Plot distribution of key categorical variables
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Age Group Distribution
data['Age_Group'].value_counts().plot(kind='bar', ax=axes[0, 0], title='Age Group Distribution')
axes[0, 0].set_xlabel('Age Group')
axes[0, 0].set_ylabel('Count')

# Gender Distribution
data['Gender'].value_counts().plot(kind='bar', ax=axes[0, 1], title='Gender Distribution')
axes[0, 1].set_xlabel('Gender')
axes[0, 1].set_ylabel('Count')

# Current Role Distribution
data['Current_Role'].value_counts().plot(kind='bar', ax=axes[1, 0], title='Current Role Distribution')
axes[1, 0].set_xlabel('Current Role')
axes[1, 0].set_ylabel('Count')

# Industry Distribution
data['Industry'].value_counts().plot(kind='bar', ax=axes[1, 1], title='Industry Distribution')
axes[1, 1].set_xlabel('Industry')
axes[1, 1].set_ylabel('Count')

# HR Experience Years Distribution
data['HR_Experience_Years'].value_counts().plot(kind='bar', ax=axes[2, 0], title='HR Experience Years Distribution')
axes[2, 0].set_xlabel('HR Experience Years')
axes[2, 0].set_ylabel('Count')

# Hide the empty subplot
axes[2, 1].axis('off')

plt.tight_layout()
plt.show()
    
```

## Effectiveness of AI Tools in HR Processes

```python

# Analyze the effectiveness of AI tools in HR processes
# Distribution of AI tools effectiveness in recruitment
ai_recruitment_effectiveness = data['AI_Tools_Effectiveness_in_Recruitment'].value_counts()

# Distribution of AI tools effectiveness in performance management
ai_performance_management_effectiveness = data['AI_Tools_Effectiveness_in_Performance_Management'].value_counts()

# Plot the effectiveness of AI tools in recruitment
plt.figure(figsize=(12, 6))
ai_recruitment_effectiveness.plot(kind='bar', title='Effectiveness of AI Tools in Recruitment')
plt.xlabel('Effectiveness Level')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Plot the effectiveness of AI tools in performance management
plt.figure(figsize=(12, 6))
ai_performance_management_effectiveness.plot(kind='bar', title='Effectiveness of AI Tools in Performance Management')
plt.xlabel('Effectiveness Level')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
    
```

## Satisfaction Levels with AI Integration in HR

```python

# Distribution of satisfaction levels with AI integration in HR
satisfaction_levels = data['Satisfaction_with_AI_in_HR'].value_counts()

# Plot the distribution of satisfaction levels
plt.figure(figsize=(12, 6))
satisfaction_levels.plot(kind='bar', title='Satisfaction Levels with AI Integration in HR')
plt.xlabel('Satisfaction Level')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
    
```

### Correlation Analysis

```python
# Convert categorical variables to numerical for correlation analysis
data['AI_Integration_Level_in_HR'] = data['AI_Integration_Level_in_HR'].astype('category').cat.codes
data['Satisfaction_with_AI_in_HR'] = data['Satisfaction_with_AI_in_HR'].astype('category').cat.codes

# Compute the correlation matrix
correlation_matrix = data[['AI_Integration_Level_in_HR', 'Satisfaction_with_AI_in_HR']].corr()

# Plot heatmap for correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Between AI Integration Levels and Satisfaction Levels')
plt.show()
```

### Trends in AI Implementation Across Different Industries

```python

# Plot trends in AI implementation across different industries
plt.figure(figsize=(12, 8))
sns.countplot(y='Industry', hue='AI_Implemented_in_HR', data=data)
plt.title('Trends in AI Implementation Across Different Industries')
plt.xlabel('Count')
plt.ylabel('Industry')
plt.legend(title='AI Implemented in HR')
plt.tight_layout()
plt.show()
```

### Distribution of AI Integration Levels by Industry

```python

# Plot the distribution of AI integration levels by industry
plt.figure(figsize=(12, 8))
sns.boxplot(x='AI_Integration_Level_in_HR', y='Industry', data=data)
plt.title('Distribution of AI Integration Levels by Industry')
plt.xlabel('AI Integration Level')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()
    
```

### Impact of AI Tools on Recruitment and Performance Management

```python
# Analyze the perceived benefits of AI tools in recruitment
benefits_recruitment = data['Benefits_of_AI_in_Recruitment'].value_counts()

# Analyze the perceived benefits of AI tools in performance management
benefits_performance_management = data['Benefits_of_AI_in_Performance_Management'].value_counts()

# Plot the perceived benefits of AI tools in recruitment
plt.figure(figsize=(12, 6))
benefits_recruitment.plot(kind='bar', title='Perceived Benefits of AI Tools in Recruitment')
plt.xlabel('Benefits')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot the perceived benefits of AI tools in performance management
plt.figure(figsize=(12, 6))
benefits_performance_management.plot(kind='bar', title='Perceived Benefits of AI Tools in Performance Management')
plt.xlabel('Benefits')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
